```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Check the dataset to remove unnessary data
nycrime = pd.read_csv("nycrime_clean.csv")
# nycrime = pd.read_csv("nycrime_clean.csv")
nycrime.info()
```

```python
# Drop unnecessary columns
nycrime_dropCOL = nycrime[['Complaint_StartDate',
                                          'Borough',
                                          'Latitude', 
                                          'Longitude' 
                                         ]]
nycrime_dropCOL.head()
```

```python
start_df = nycrime_dropCOL.dropna()
```

```python
# Check the dataset to remove unnessary data
nycrime_date = pd.read_csv("weekday_month_holiday.csv")
nycrime_date.head()
```

```python
nycrime_date['date'] = pd.to_datetime(nycrime_date.date)
nycrime_date.head()
```

```python
nycrime_date['date'] = nycrime_date['date'].dt.strftime('%m/%d/%Y')
nycrime_date.head()
```

```python
# Check the dataset to remove unnessary data
nycrime_zipcode = pd.read_csv("nycrime_zipcode_test.csv",encoding='latin1')
nycrime_zipcode.info()
```

```python
nycrime_new = nycrime_zipcode[['Latitude', 'Longitude', 'Zipcode']]
nycrime_new.head()
zip_df = nycrime_new.dropna()
```

```python
zip_df.head()
```

```python
rounded_zip_df = zip_df.round({"Latitude":4,"Longitude":4})
```

```python
start_df.head()
```

```python
rounded_start_df = start_df.round({"Latitude":4,"Longitude":4})
```

```python
final_df = pd.merge(rounded_start_df, rounded_zip_df,how='left', left_on=['Latitude','Longitude'], right_on=['Latitude','Longitude'])
```

```python
final_dropped = final_df.dropna()
```

```python
final_dropped.rename(columns={'Complaint_StartDate': 'date'}, inplace=True)
final_dropped.head()
```

```python
ny_test = pd.merge(final_dropped, nycrime_date, on='date')
ny_test
```

```python
# Save clean dataFrame as CSV file
ny_test.to_csv('nycrime_zipcode_ML_post_age.csv')
```
