# New York City Crime Data Cleansing

```python
# Import dependencies
import pandas as pd
import datetime
```

```python
# Check the dataset to remove unnessary data
nycrime = pd.read_csv("NYPD_Complaint_Data_Historic.csv")
nycrime.info()
```

# Filtering Jurisdiction to NYPD

```python
# Check all of the NYC policing jurisdictions
nycrime['JURIS_DESC'].value_counts()
```

```python
# Filter jurisdictions to NYPD only
nycrime.set_index('JURIS_DESC',inplace=True)
nycrime_NYPD = nycrime.loc['N.Y. POLICE DEPT']
nycrime_NYPD.index.value_counts()
```

```python
# Remove the jurisdiction index column by resetting the index
nycrime_NYPDclean = nycrime_NYPD.reset_index(drop=True)
nycrime_NYPDclean
```

# Dropping and Renaming Columns

```python
# Drop unnecessary columns
nycrime_dropCOL = nycrime_NYPDclean.drop(['CMPLNT_NUM', # RNG complaint ID# not needed. 
                                          'ADDR_PCT_CD', # precinct# not needed.
                                          'RPT_DT', # Police report date not needed.
                                          'JURISDICTION_CODE', # Related to JURIS_DESC. Not needed.
                                          'PATROL_BORO' # BORO_NM is a better location indicator.
                                         ], axis=1)
```

# Filtering Dates to 2006 to 2017

```python
# Check starting date values. Verify datatype of starting date as string.
nycrime_renameCOL['Complaint_StartDate'].head()
```

```python
# Convert date string into datetime
nycrime_renameCOL["dateTime_start"]=pd.to_datetime(nycrime_renameCOL['Complaint_StartDate'], errors='coerce', format='%m/%d/%Y')
```

```python
# Verify datatype of starting date as datetime
nycrime_renameCOL["dateTime_start"].head()
```

```python
# Filter date to start at 2006
nycrime_filterDate = nycrime_renameCOL[(nycrime_renameCOL['dateTime_start'] >= '2006-01-01')]
```

```python
# Verify range of filtered starting dates
# nycrime_FilterDATE['dateTime_start']
# nycrime_FilterDATE['dateTime_start'].value_counts()
nycrime_filterDate['dateTime_start'].describe()
```

# Filtering Age Groups

```python
# Check suspect age values
nycrime_filterDate['Suspect_Age'].value_counts()
```

```python
# Check victim age values
nycrime_filterDate['Victim_Age'].value_counts()
```
